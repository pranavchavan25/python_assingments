```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# sample data
tips = sns.load_dataset("tips")

# Histogram with KDE(kernal density estimation) using seaborn
sns.histplot(tips["total_bill"],bins = 10,kde=True, color="skyblue" )
plt.title("Distribution of total bill")
plt.xlabel("Total Bill")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python
sns.kdeplot(tips["total_bill"],fill = True,color = "orange")
plt.title("kde Plot of Total Bill")
plt.xlabel("Total Bill")
plt.show()
```

```python
sns.displot(data = tips, x = "total_bill", hue = "sex", kind = "kde",fill="True")
plt.show()
```

# Catogroial plot(Bar graph)

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")

# Average total bill for each day
sns.barplot(x = "day",y = "total_bill",data= tips,hue="day",palette = "pastel")
plt.title("Average  total billper day")
plt.show()
```

# HeatMap

```python
# sample data: student score in 3 subjects
score = pd.DataFrame({
    "Maths":[85,78,92],
    "Science":[88,76,95],
    "English":[90,80,85]
},index=['Student A','Student B','Student C'])

# creating heatmap
sns.heatmap(score, annot = True, cmap= 'Greens')    # annto = to show the value in that block
plt.title('Student Scores Heatmap')
plt.show()
```
